### Fitting

The statsmodels function for fitting LINEAR MODELS for regression is `smf.ols()`.

The `smf.ols()` function works slightly differently than how the sklearn works. The object is INITIALIZED with a DATA SET!

You then FIT the object with the `.fit()` method BUT the data has already been provided!

The model we want to FIT is specified via the `formula` arguement to `smf.ols()`.

The FORMULA is a STRING!

`'<OUTPUT> ~ <INPUT>'`

By default, the FORMULA assumes an INTERCEPT **MUST BE ESTIMATED** !

The resulting FITTED object will include all COEFFICIENT ESTIMATES, STANDARD ERRORS, CONFIDENCE INTERVALS, and most PERFORMANCE METRICS.

```python
lm_fit = smf.ols( formula='y ~ x' , data=df ).fit()
```

```python
type( lm_fit )
```

```python
dir( lm_fit )
```

The `.summary()` is a useful METHOD for SUMMARIZING the linear model FIT!

```python
lm_fit.summary()
```

' I prefer to PRINT the summary ' - Professor

```python
print(lm_fit.summary() )
```

As a reminder, the FORMULA of `'y ~ x'` stands for:

$$\mu = \beta_0 + \beta_1 \times x $$

In the summary results, the name `x` is $$\beta_0$$, and the name `intercept` is $$\beta_1$$

The `.params` attribute stores the COEFFICIENT ESTIMATES! These are the estimates that MINIMIZE the sum of squared errors!

```python
lm_fit.params
```

```python
lm_fit.bse
```

The standard erroor on the estimate is contained the `.bse` attribute.

Essentiallly, we can use the $\pm1$ and $\pm2$ standard error intervalls to know how TRUSTWORTHY the estimate is!

The most common standard error interval is the $\pm2$ standard error interval becasue that is approximately equal to the 95\% CONFIDENCE INTERVAL!

```python
lm_fit.params - 2 * lm_fit.bse
```

```python
lm_fit.params + 2 * lm_fit.bse
```

We are 95\% confident that the coefficient is somewhere between -2 and -1.3

The actual 95\% confidence interval is a slioghtly more difficult formula. But when the sample size is high our approximation using the Gaussian becomes really good. If you want the EXACT 95\% confidence interval, then you can extract it using the `.conf_int()`.

```python
lm_fit.conf_int()
```

```python
type( lm_fit.conf_int())
```

```python
lm_fit.conf_int().columns
```

Let's change the column names to the CONFIDENCE INTERVAL LOWER BOUND and the CONFIDENCE INTERVAL UPPER BOUND.

```python
lm_fit.conf_int().\
rename(columns={0: 'ci_lwr', 1: 'ci_upr'})
```

I will now extract the 95\% confidence interval bounds and store the ESTIMATE with the STANDARD ERROR with the bounds.

```python
coef_fit_info = lm_fit.conf_int().\
rename(columns={0: 'ci_lwr', 1: 'ci_upr'})
```

```python
coef_fit_info
```

```python
coef_fit_info['estimate'] = lm_fit.params
```

```python
coef_fit_info['estimate_se'] = lm_fit.bse
```

```python
coef_fit_info
```

```python
coef_fit_info.index
```

We can extract the p-values for the coefficients using the `.pvalues` attribute.

```python
lm_fit.pvalues
```

The p-value is described as something really magical and important. It is used to identify **STATISTICALLY SIGNIFICANT** inputs.

A STATISTICALLY SIGNIFICANT input is one that has a p-value typically less than **0.05**

```python
lm_fit.pvalues < 0.05
```

But the p-value is nothing magic or special, in fact it can be completely described using a simple graphic.

Let's visualize the coefficient ESTIMATES as markers in a scatter plot and the CONFIDENCE INTERVALS as error bars.

We need to use the matplotlib axis object `.errorbar()` method to visualize the CI.

```python
coef_fit_info
```

```python
lm_fit.params
```

```python
lm_fit.bse
```

```python
lm_fit.conf_int()
```

The `.errorbar()` method wants us to provide the VALUE of the ERROR around the MARKER rather than the LIMITS of the INTERVAL.

Therefore, I typically create the following **COEFFICIENT PLOT** visualization using the approximation based on 2 times the standarad error.

```python
fig, ax = plt.subplots()

### use the errorbar method to show the estimates as markers and CI as errorbars
### create a HORIZONTAL errorbar
ax.errorbar(y = coef_fit_info.index,
            x = coef_fit_info.estimate,
            xerr = 2 * coef_fit_info.estimate_se,
            fmt='o', color='k', ecolor='k', elinewidth=2, ms=8)

### INCLUDE a VERTICAL REFRENCE LINE at 0
ax.axvline(x=0, linestyle='--', linewidth=4, color='grey' )

### set axis labels
ax.set_xlabel('coefficient value')
ax.set_ylabel('coefficient name')

plt.show()
```

The confidence or trustworthiness off the SIGN (NEGATIVE or POSTITIVE) of the SLOPE is typically expressed via the p-value

```python
lm_fit.pvalues
```

I will typically use the fitted statsmodels object directly to create the above figure.

```python
fig, ax = plt.subplots()

### the 95% CI approximation around the estimate
ax.errorbar( y=lm_fit.params.index,
             x=lm_fit.params,
             xerr = 2 * lm_fit.bse,
             fmt='o', color='k', ecolor='k', elinewidth=2, ms=10)

### reference line at 0
ax.axvline(x=0, linestyle='--', linewidth=4, color='grey')

### axis labels
ax.set_xlabel('coefficient value')

plt.show()
```

However, if you want the EXACT 95% confidence intervals, then we need to do some extra manipulations using the `.conf_int()` method.

```python
lm_fit.conf_int()
```

```python
coef_fit_info
```

We need to calculate the lower change on the error and the upper change on the error bar around the average.

```python
coef_fit_info['lwr_err'] = coef_fit_info.estimate - coef_fit_info.ci_lwr
coef_fit_info['upr_err'] = coef_fit_info.estimate - coef_fit_info.estimate
```

```python
coef_fit_info
```

```python
fig, ax = plt.subplots()

ax.errorbar( y = coef_fit_info.index,
             x = coef_fit_info.estimate, 
             xerr = [ coef_fit_info.lwr_err, coef_fit_info.upr_err ],
             fmt='o', color='k', ecolor='k', elinewidth=3, ms=10 )

ax.axvline(x=0, linestyle='--', linewidth=5, color='grey')

ax.set_xlabel('coefficient value')

plt.show()
```

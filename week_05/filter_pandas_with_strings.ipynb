### Filtering Rows:

Filtering refers to `selecting` rows based on the Conditional Test!

```python
baseball_df.loc[ baseball_df.wins > 65, : ]
```

```python
baseball_df.loc[ baseball_df.Team == 'Pirates', : ]
```

We also saw how to use the `or` operator, `|`, to find all rows where the value equals A or B.

Or, the value is ONE OF those presented!

```python
baseball_df.loc[ (baseball_df.Team == 'Cardinals') | ( baseball_df.Team == 'Brewers'), : ]
```

The `==` operator combined with the `|` operator iss correct to use... but it does not SCALE well.

For example, if we needed to check for 10 possible values, we would need to type in 10 different conditions.

Instead, we can use the `.isin()` operator to streamline the `|` operator!

```python
baseball_df.loc[ baseball_df.Team.isin(['Cardinals', 'Brewers']), :]
```

```python
baseball_df.loc[ baseball_df.Team.isin(['Cardinals', 'Brewers', 'Pirates']), :]
```

The `.isin()` method can alo be applied to numbers.

```python
baseball_df.loc[ baseball_df.losses.isin([70, 90]), :]
```

The `.isin()` operator is especially useful for string filtering!

```python
top_teams = baseball_df.loc[ baseball_df.games_back < 10, 'Team'].copy().tolist()
```

```python
top_teams
```

```python
baseball_df.loc[ baseball_df.Team.isin(top_teams), :]
```

### String Pattern Matching:

```python
baseball_df.loc[ baseball_df.City == 'Pittsburgh', :]
```

But what if I didn't feel like typing out the entire string of `Pittsburgh` ?

```python
baseball_df.loc [ baseball_df.City == 'Pitt', :]
```

What if we had a typo in our search?

```python
baseball_df.loc[ baseball_df.City == 'Pittsburg', :]
```

Instead, we could instead focus on a PATTERN. The `.str.contains()` method searches for a PATTERN within the string!

```python
baseball_df.City.str.contains('Pitt')
```

```python
baseball_df.City
```

```python
baseball_df.loc[ baseball_df.City.str.contains('Pitt'), :]
```

We can even apply the PATTERN search to a single character!

```python
baseball_df.loc[ baseball_df.City.str.contains('P'), :]
```

```python
baseball_df.loc[ baseball_df.City.str.contains('C'), :]
```

Notice what happened here, we need to be careful how freely we use the operator in a large list. If the list was 20,000 objects long and we searched something as broad as `'C'`, we might get a ton of returning values that fit the description but not entirely what we might be searching for. 

*It is recommended that we specify our search as much as possible in larger data libraries.*

It is best used to search for non-letter characters. To find a peried in a string, we need to search for the pattern `\\.`.

```python
baseball_df.loc[baseball_df.City.str.contains( '\\.'), :]
```

Let's say that we wanted to narrow our list to teams that are a combination of words (include a space in their city name). Example: New York, New Orleans, etc.

```python
baseball_df.loc[baseball_df.City.str.contains( ' '), :]
```

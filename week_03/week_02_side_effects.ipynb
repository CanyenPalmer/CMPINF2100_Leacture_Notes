The reason for the strange behavior is that Python has **side effects** from many actions.

A side effect: is something that occurs as part of defining an object or executiing an action.

Python does not actually *attach* or *use* an object to define a new object.

Instead, Python **points** to the original object.

```python
snacks_dict['candy bars']
```

```python
snacks_list[0]
```

```python
snacks_dict_1
```

```python
snacks_dict_1['candy bars']
```

```python
snacks_1
```

```python
added_snacks
```

We must be careful in Python when using existing objects (variables, assigned) to define new objects!

Because changing the original object may have a side effect that changes later objects in our programming!

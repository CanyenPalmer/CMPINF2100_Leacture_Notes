#### For-Loops:

```python
for n in range(7):
    # Inside te body of the for loop
    # Actions occur for the Current value of the iterating variable
    print(n)
```

```python
for n in range(7):
    # Inside the body off the for-loop
    # Still inside the body
    print(n)
    # Inside the body of the for-loop
    print(n)
# We can use ctrl + `[` to reverse indent 
# Since we are not indented, we are no longer iterating
print(n)
print(n*100)
# Still outside of the for-loop
```

```python
for n2 in range(7):
    # Inside the for-loop
# Outside of the for-loop
print(n2)
```

The iterating variable is only created when something is completed inside of the body of the for-loop

This means that it must be iterated in the loop before it can be printed.

```python
for n2 in range(7):
    # Inside the for-loop
    print(3)
print(n2)
```

Notice how we didn't specifically iterate or call `n2` inside of the for-loop. It worked anyways since something was iterated at least.

```python
for n3 in range(7):
    # the for-loop expects some kind of action!
    n3 + 1
    # we are not printing, but we are adding a integer to the variable
# outside of the for-loop
print(n3)
```

```python
for nn in range(7):
    print("Iterating variable value is %d" % (nn) )
```

```python
for mm in range(7):
    print("Old iterating variable, nn, is: %d While current iterating variable, mm, is %d" % (nn, mm) )
```

`nn` is outside of the for loop and not changing while `mm` is inside of the for-loop and changing on each iteration.

```python
for zz in range(7):
    print("%d + %d + %d = %d" % (nn, mm, zz, nn+mm+zz) )
```

#### Nested For-loops:

We can iterate over 2 sequences at the same time! Nested for-loops consist of outer and inner for-loops!

```python
for o1 in range(3):
    # this is the outer for-loop
    print("Outer iterationg number: %d" % (o1) )
    print ("--##--")
    for i1 in range(5):
        # this is inside the inner for loop
        # it looks like a futher sub-section since we are indented more
        print("Inner iteration number: %d" % (i1) )
    # we are back within the outer for-loop
    print("--$$--")
```

```python
for o2 in range(5):
    # outer for-loop iterates over an integer:
    print("Outer iteration number: %d" % (o2) )
    for i2 in ['a', 'b', 'c', 'd']:
        # inner for-loop:
        print(o2*i2)
    # back in the outer for-loop
    print("--$$--")
```

Whenever we multiply an `integer` with a `string`, it repeats the string the number of times that the integer assigns. So the first block as nothing since we asked the model to repeat the `string` '0' times.

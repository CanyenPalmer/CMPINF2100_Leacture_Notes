## Controlling randomness

We need  way to repeat the random result!

We can control randomness in computer programs by setting the random **seed**!

The `random.seed()` function allows us to set or define the seed that controls the random number generators!

```python
random.seed(2100)
```

```python
random.random()
```

```python
random.random()
```

Let's include all three function calls in a single cell.

```python
random.seed(2100)

print(random.random())

print(random.random())
```

Note: Think that the **seed** is planted and it will determine how the **tree** will grow. That is why whenever we continuously recall the same cell with the same seed, it falls back to the first initial random cell. 

Let's demonstrate this one more time. 

Call `random.random()` within list!

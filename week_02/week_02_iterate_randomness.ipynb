#### List Comprehension

The list comprehension is a streamlined way of managing iteration. 

List comprehensions require that we completely understand the 3 basic elements of iteration:

- the action to repeat
- the iterating variable that changes
- the sequence to iterate over

In code, the syntax is:

`<output list> = [ <action> for <variable> in <sequence> ]`

----------------------------------------------------------------------------------


What is the action that we are repeating?

- we are calling `random.random()` many times

What is the iterating variable?

- we are calling it `n` but it doesnt really do anything here...

What is our sequence to iterate over?

- The sequence is generated by `range()`

```python
random.seed(2100)

y = [ random.random() for iterating_variable in range(11) ]
```

```python
type(y)
```

```python
len(y)
```

```python
y
```

```python
y == xu_c
```

The iterating variable is not really important in our current application, but most off the time it is very important. But when it is not needed in a common practice, it makes more since to use the underscore variable '_' as the iterating variable.

```python
random.seed(2100)

y2 = [ random.random() for _ in range (11) ]
```

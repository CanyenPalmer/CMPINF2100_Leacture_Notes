## CMPINF 2100 Week 07 | Visualize Continuous -> Continuous Relationships via Trend Plots

The `Trend Plot` is an extension that adds further **quantitative** information about the relationship between two continuous variables.

### Import Modules:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.info()
```

### Scatter Plot

Show the relationship between two continuous variables.

```python
sns.relplot(data=penguins, x='flipper_length_mm',y='bill_depth_mm')

plt.show()
```

We saw that by *grouping* or *conditioning* on a categoriacal variable caused our interpetation of the scatter plot changed!

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='species')

plt.show()
```

### Trend Plots

We can add **Trend Lines** to visualize the linear relationship between the two continuous variables. 

TREND LINES correspond to the **Best Fit Line** between two variables. 

Later on... we will learn how this line is created. For now, let's visualize it and let the line or TREND help us interpret the relationship!

The TREND is NOT created with `sns.relplot()` function. 

Instead, the TREND is created with the FIGURE LEVEL function `sns.lmplot()`.

```python
sns.lmplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm')

plt.show()
```

We are viewing an ANTI or NEGATIVE relationship between the two variables!

But, what happens if we GROUP or CONDITION on a categorical variable?

Let's color via the `hue` argument by `species`.

```python
sns.lmplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species')

plt.show()
```

Given or GROUPING BY or CONDITIONING ON the `species` changes our opinion about the relationship between the 2 continuous variables. Grouping by `species` reveals that there is a POSITIVE relationship between the two!

- If the flipper length is bigger the bill debth should be bigger, via a larger penguin should genetically default to larger variables.

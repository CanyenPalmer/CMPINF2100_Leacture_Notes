### Continuous to Continuous Relationships

The fundamental plot for showing the relationship between 2 continuous variables is the **Scatter plot**.

In seaborn, the FIGURE LEVEL function for showing RELATIONSHIPS is the `sns.relplot()` function.

The `x` axis is associated with a continuous variable and the `y` axis is associated with a second continuous variable.

```python
sns.relplot(data = penguins, x = 'flipper_length_mm', y='bill_depth_mm')

plt.show()
```

```python
penguins.info()
```

```python
sns.relplot(data=penguins, x='body_mass_g', y='bill_length_mm')

plt.show()
```

The underlying AXIS level function called by the FIGURE LEVEL `sns.relplot()` is the `sns.scatterplot()` function.

```python
fig, ax = plt.subplots()

sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'bill_length_mm', ax=ax)

plt.show()
```

In Summary:

* The scatterplot shows the relationship between two continuous variables
* However, a scatterplot can show more than just **2** variables
* A common practice is to include a categorical variable via color, we can associate a categorical variable with a `hue` argument.

Let's return to our first plot.

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm')

plt.show()
```

But we have categorical variables, object data types, in the `penguins` data set!

```python
penguins.info()
```

What happens if we group by a categorical variable?

Otherwords, we can study the relationship between 2 variables changes across categories! This means we can measure if the categorical variable is changing!

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species')

plt.show()
```

Conditioning or Grouping by a categorical variable can **change** our opinion on the relationship between two continuous variables!

Instead of coloring by the categorical variable, we can also use FACETS to SEPERATE the groups.

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm', col='species')

plt.show()
```

You can combine both `hue` and `col` facets to get facets and colors. 

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species', col='species')

plt.show()
```

If you do not color by the same variables as the **facet**, then you can use different categorical variabless for color and facet!

You can group the relationships on 2 seperate and distinct categorical variables. You are grouping by the COMBINATION of 2 categorical variables!

The combination on `species` and `island` on the relationship!

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm', hue='species', col='island')

plt.show()
```

Here now, we have 2 continuous variables on 2 different categories. This means that we have 4 total variables that are being studies here!

We can add yet another categorical variable through row facets!

```python
penguins.info()
```

If we use row and column facets and associate `hue` with a categorical variable, we are effectively GROUPING the relationship between 2 continuous variables on the combination of 3 categorical variables!

```python
sns.relplot(data=penguins, x='flipper_length_mm', y='bill_depth_mm', 
            hue='sex', col='island', row='species')

plt.show()
```

Remember to always check the COMBINATION between categoricals BEFORE looking at the relationships between continuous variables!

```python
sns.catplot(data=penguins, x='island', hue='species', kind='count')

plt.show()
```

In Review:

* Scatter plots measure continuous to continuous relationships
* we are able to include additional variables via `hue` argument through column facets
* you must always consider the sample sizes before comparing the relationships between the continuous variables (this can save time if there is no immediate correlation or flag attention if there is a lot)

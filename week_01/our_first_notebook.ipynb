VERY IMPORTANT: the beggining index is INCLUSIVE but the ending is EXCLUSIVE!

```python
my_string[0:3]
```

```python
len(my_string[0:3])
```

```python
my_string[3]
```

```python
my_string[1:3]
```

## Iterate with for-loops

A for-loop allows us to APPLY an action or EXECUTE a procedure over and over again.

This way, we do not have to manually apply each action, or copy-paste (which is dangerous in coding)

A for-loop iterates over a SEQUENCE by changing the value of an ITERATING VARIABLE along that sequence.

```python
for a_golfer in the_golfers:
    print(a_golfer)
```

```python
%whos
```

```python
the_golfers[-1]
```

```python
a_golfer
```

A common way to create or define a for-loop is with a `range()` function.

```python
range(4)
```

```python
print(0)
print(1)
print(2)
print(3)
```

```python
for n in range(4):
    print(n)
```

```python
%whos
```

The `range()` function is often used to GENERATE a SEQUENCE of INTEGERS to SUBSET lists!

```python
for m in range(4):
    print(the_golfers[m])
```

```python
for a_golfer in the_golfers:
    print(a_golfer)
```

We do not need to HARD CODE the 4 in `range()`. We can use the `len()` function to give us the LENGTH of the list!

```python
len(the_golfers)
```

```python
for n in range(len(the_golfers)):
    print(the_golfers[n])
```

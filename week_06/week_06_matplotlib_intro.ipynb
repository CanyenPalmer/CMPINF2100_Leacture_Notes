### Overview:

`matplotlib` has 2 main styles of programming:

1) the first or classsic is the STATEFUL style which is based on MATLAB.
2) the second approch is the object oriented style approach

### Stateful or MATLAB Style

* step 1: initiate the figure
* step 2: specify the subplot that containss the data
* step 3: populate the data within the plot (plot the data)
* step 4: show the figure

Let's begin with step one and work through each step, one at a time.

```python
### Step 1 - initiate the figure

plt.figure()
```

Let's initiate a single figure that has 2 subplots. One shows the SINE WAVE and teh other shows a COSINE WAVE

```python
plt.figure()
### define 2 subplots
plt.subplot(2,1,1)
plt.plot(np.sin(np.linspace(0,10,100)))

plt.subplot(2,1,2)
plt.plot(np.cos(np.linspace(0,10,100)))

plt.show()
```

Examine this again, but only populate data in the second subplot

```python
plt.figure()

plt.subplot(2,1,1)
plt.subplot(2,1,2)

plt.plot(np.cos(np.linspace(0,10,100)))

plt.show()
```

We could instead use 1 row and 2 columns of subplots

```python
plt.figure()

plt.subplot(1,2,1)
plt.plot(np.sin(np.linspace(0,10,100)))

plt.subplot(1,2,2)
plt.plot(np.cos(np.linspace(0,10,100)))

plt.show()
```

### Object Oriented Style

Many aspects of the Object Oriented Style figures are the same as the STATEFUL figures. But the way we access or apply functions to them are different!

* step 1: Define the figure and axes objects - Create Variables
* step 2: Plot the data within the axes using methods associated with the axes objects!
* step 3: Modify the axes object attributes if needed
* step 4: Modify the figure object attributes if needed
* step 5: show the figure with `plt.show()`

The motivation for the object oriented approach is that it is the STATEFUL approach works with **hidden objects**.

Define variables or objects for the *figure* and *axis* that we are modifying.

The common names for these objects are `fig` and `ax`.
